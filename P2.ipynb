```python
from collections import deque

# DFS Function
def dfs(cap1, cap2, target):
    stack = [((0, 0), [])]  # Initial state (Jug1, Jug2)
    visited = set()

    while stack:
        (jug1, jug2), path = stack.pop()

        if (jug1, jug2) in visited:
            continue

        visited.add((jug1, jug2))
        path = path + [(jug1, jug2)]

        # Goal condition
        if jug1 == target or jug2 == target:
            return path

        # Possible operations
        next_states = [
            (cap1, jug2),                          # Fill Jug1
            (jug1, cap2),                          # Fill Jug2
            (0, jug2),                             # Empty Jug1
            (jug1, 0),                             # Empty Jug2

            # Pour Jug1 -> Jug2
            (
                jug1 - min(jug1, cap2 - jug2),
                jug2 + min(jug1, cap2 - jug2)
            ),

            # Pour Jug2 -> Jug1
            (
                jug1 + min(jug2, cap1 - jug1),
                jug2 - min(jug2, cap1 - jug1)
            )
        ]

        for state in next_states:
            if state not in visited:
                stack.append((state, path))

    return None

# Main Program
jug1_capacity = 4
jug2_capacity = 3
target = 2

solution = dfs(jug1_capacity, jug2_capacity, target)

if solution:
    print("Steps to get", target, "litres:\n")
    step = 0
    for state in solution:
        print("Step", step, ":", state)
        step += 1
    print("\nGoal Reached!")
else:
    print("No solution found.")
print("Lavkush T048")
```

```python
import heapq

graph = {
    'Mumbai': [('Pune', 150), ('Hyderabad', 710)],
    'Pune': [('Mumbai', 150), ('Bangalore', 840)],
    'Hyderabad': [('Mumbai', 710), ('Bangalore', 570)],
    'Bangalore': [('Pune', 840), ('Hyderabad', 570)]
}

def ucs(start, goal):
    queue = [(0, start, [start])]
    visited = []

    while queue:
        cost, city, path = heapq.heappop(queue)

        if city == goal:
            return cost, path

        if city not in visited:
            visited.append(city)

            for neighbor, distance in graph[city]:
                heapq.heappush(
                    queue,
                    (cost + distance, neighbor, path + [neighbor])
                )

cost, path = ucs("Mumbai", "Bangalore")

print("Least-cost path:")
print("Path:", " -> ".join(path))
print("Total Distance:", cost, "km")
print("Lavkush T048")
```
